```python
# Import pandas
import pandas as pd

# Import Twitter data as DataFrame: df
df = pd.read_csv('tweets.csv')

df.head()
```

```python
# Extract column from DataFrame: col
col = df['lang']
#col = df[lang]
print(col)
```

```python
# Initialize an empty dictionary: langs_count
langs_count = {}


langs_count['bb'] = 1
print(langs_count)
print(langs_count.keys())
print(langs_count.values())
```

```python
langs_count = {}
for i in col:
    
    if langs_count[i-1]==langs_count[i]:
        langs_count[i] += 1
print(langs_count)
print(langs_count.keys())
print(langs_count.values())
```
